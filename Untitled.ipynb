```python
from IPython.display import display 
from sklearn.feature_extraction.text import CountVectorizer

from sklearn import svm
```

BAG OF WORDS APPROACH
=====================

**Books and Clothing Classification: ==>>**
It the easies way to take sentence and convert it to numerical representation.
It extra all unique words in series sentences to create a dictionary of words; and matches these sentences over the dictionary `(1 if the sentence has the word and 0 if the sentence don't have the word)`

**LIMITATION OF BAG OF WORDS: ==>>**
If we have a word that's not in our training uttrances the model will not know how to handle it. 
Its great on the words it has trained with but if you input a new word then it will fail.

```python
class Category:
    BOOK = "books"
    CLOTHING = "clothing"

train_x = ["I love the book", "This is a great book", "The fit is great", "I love the shoes"]
train_y = [Category.BOOK, Category.BOOK, Category.CLOTHING, Category.CLOTHING]
```

```python
vectorizer = CountVectorizer(binary=True, ngram_range=(1, 2))
```

```python
vectors = vectorizer.fit_transform(train_x)
```

```python
display(vectorizer.get_feature_names())
vectors.toarray()
```

```python
clf_svm = svm.SVC(kernel="linear")

clf_svm.fit(vectors, train_y)
```

```python
test_x = vectorizer.transform(["I like the story"])
clf_svm.predict(test_x)
```

WORD VECTORS APPROACH
=====================

**Books and Clothing Classification: ==>>**
Convert words in to numerical vectors.
`The big goal of word vector is to covert text in to a numerical vector that capture some of the semantic meaning in a vector space that you are mapping.`

They are many different ways to achieve this. But the most common is `"word2vec"`. There are two ways to train word2vec.
* continue bag of words
* skipgram

What this do is look at a window of text to get the semantics.
* spacy library is good for `word2vec`.

**LIMITATION OF WORD VECTORS: ==>>**
* When the categories and the sentences are too long, the meaning of words turns to loss value.
* Sometimes they are not kind of precieved as when using bag of words.
* For words with more than one meaning it get messes, trying to capture their meaning during training and some of those meanings are probabily loss in the process

WORD VECTOR are very cool and powerful but they don't solve everything.

```python
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
docs = [nlp(text) for text in train_x]
display(docs)

train_x_word_vector = [x.vector for x in docs]
train_x_word_vector
```

```python
clf_svm_mv = svm.SVC(kernel="linear")

clf_svm_mv.fit(train_x_word_vector, train_y)
```

```python
test_x = ["I love the story"]
test_docs = [nlp(text) for text in test_x]

test_x_word_vector = [x.vector for x in test_docs]
clf_svm_mv.predict(test_x_word_vector)
```
