```python
from IPython.display import display

import re
import requests
from bs4 import BeautifulSoup

import spacy
```

```python
URLS = ["https://link.springer.com/article/10.1057/rpm.2010.39", 
        "https://www.cpajournal.com/2019/04/08/new-revenue-recognition-guidance-and-the-potential-for-fraud-and-abuse/",
        "https://www.investopedia.com/ask/answers/09/how-companies-calculate-revenue.asp"]

page = requests.get(URLS[2])
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
for section in soup.find_all("section"):
    for p in section.find_all("p"):
        p_tags = p.text.lower()
        
        print(p.text)
        for match in re.finditer("revenue", p_tags):
            print(match.start(), match.end(), "\t", "REVENUE")
            print("\n")
    print("\n")            
```

```python
for section in soup.find_all("div", class_="articleSection"):
    for p in section.find_all("p"):
        p = p.text.lower()
        
        print(p)
        for match in re.finditer("revenue", p):
            print(match.start(), match.end(), "\t", "REVENUE")
            print("\n")

    print("\n")
```

```python
for article in soup.find_all("article"):
    for p in article.find_all("p"):
        p_tags = p.text.lower()
        
        print(p.text)
        for match in re.finditer("revenue", p_tags):
            print(match.start(), match.end(), "\t", "REVENUE")
            print("\n")
            
    print("\n")
    
```
